```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import scipy.stats as stats
```

```python
def modexponential(x,tau, K,c):
    y=-K*np.exp(tau*x)
    y=y+c
    y=y**2
    y=np.sqrt(y)
    return(y)
```

```python
y=[10.51,9.1,7.76,6.55,5.46,4.42,3.44,2.54,1.73,0.97,0.53,0.74,1.35,1.9,2.41,2.9,3.37,3.81]
x=np.arange(0.002,19*0.002,0.002)
y_err=(0.07,0.11,0.06,0.03,0.03,0.03,0.03,0.03,0.03,0.03,0.06,0.03,0.02,0.02,0.03,0.02,0.03,0.03)
y_err=np.array(y_err)
plt.errorbar(x,y,y_err, fmt='r.', label='Data')
fitpar=scipy.optimize.curve_fit(modexponential,x,y,[-5,18,6])
T1=1/fitpar[0][0]
T1_err=fitpar[1][0][0]/(fitpar[0][0]**2)
print('Lifetime')
print(str(T1) + '+/-' + str(T1_err))
#yguess=modexponential(x,-5,18,6)
yfit=modexponential(x,*fitpar[0])
plt.plot(x,yfit,label='Fit')
#plt.plot(x,yguess)
residuals=y-yfit
pulls=residuals/y_err
chisquare=np.sum(pulls**2)
dof=len(pulls)-len(fitpar[0])
print('Reduced chi squared')
print(chisquare/dof)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.legend()
```

```python
plt.figure()
plt.scatter(x,residuals)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.show()
plt.figure()
plt.scatter(x,pulls)
plt.xlabel('Time (s)')
plt.ylabel('Standard deviations (sigma)')
plt.show()
```
